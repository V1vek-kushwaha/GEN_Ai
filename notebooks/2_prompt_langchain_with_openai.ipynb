## Dynamic Prompts

```python
from langchain_google_genai import GoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = GoogleGenerativeAI(model="gemini-2.5-flash")
out = StrOutputParser()
 

 # my selft chian node

def transform_case(result:str):
    return result.upper()

prompts= ChatPromptTemplate.from_messages([
    {"role":"system","content":"You are a translator and translate input in {language}"},
    {"role":"user","content":"{quary}"}

])

# final_prompts = prompts.format_messages(language = "hindi", quary = "i love python and js")

# res = llm.invoke(final_prompts)
# res


# res = llm.invoke(prompts.invoke({"language":"Hindi","quary":"i love you !"}))
# res

chains = prompts | llm | out | transform_case
res = chains.invoke({"language":"Hinglish","quary":"i love langchain !"})
print(res)
```
