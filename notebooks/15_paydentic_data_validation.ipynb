```python
from pydantic import BaseModel,Field
from typing import List , Annotated
```

```python
class User1(BaseModel):
    name : str
    age : int
   
```

```python
user = User1(name="vivek",age=56)
user
```


### Field Validation

```python
class Product(BaseModel):
    title:str = Field(max_length=12)
    des :str = Field(min_length=20)
    price: int = Field(ge=0 , lt=100)
    
```

```python
mobile = Product(title="samsung",des="this is multinatination mobile company with there flgship expriance",price=16)
cooler = Product(title="samsung",des="this is multinatination coller company with there flgship expriance",price=13)
```

```python
class Order(BaseModel):
    id:str = Field(max_length=12 , min_length=5)
    product:Product
    
```

```python
order = Order(id="1223sf",product=mobile)
order
```

```python
class Cart(BaseModel):
    id:str = Field(min_length=6 ,max_length=12)
    items: List[Product]
    comment:str = Field(default="no comment")
```

```python
cart = Cart(id="aajdflkahjdf",items=[mobile,cooler],comment="this is my cart")
```

```python
cart
```

```python
for p in cart.items:
    print(p.title)
```



#Annotated

```python
common_Validator = Annotated[List[Product],Field(default=[])]
```

```python
class Cart(BaseModel):
    id:str = Field(min_length=6 ,max_length=12)
    items: common_Validator
    comment:str = Field(default="no comment")
```

```python
class Orders(BaseModel):
    products:common_Validator
```

```python
ord = Orders(products=[mobile,cooler])
```
