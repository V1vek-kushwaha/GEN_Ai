```python
from dotenv import load_dotenv
load_dotenv()
from langchain_core.tools import tool
from langchain_groq import ChatGroq
from langchain.agents import create_agent

llm = ChatGroq(model="openai/gpt-oss-120b")

@tool
def add_number(a:int , b:int):
    """
        it will return the sum of two numbers

        Args:
            a : number one
            b : number two
    """
    return a + b

add_number.invoke({
    "a":12,
    "b":13
})

@tool
def sub_number(a:int , b:int):
    """
        it will return the subtract of two numbers

        Args:
            a : number one
            b : number two
    """
    return a - b

add_number.invoke({
    "a":12,
    "b":13
})




agent = create_agent(
    model=llm,
    tools=[add_number,sub_number],
    system_prompt="you are a math teacher and  always use tools for calculations"
    )


res = agent.invoke({"messages":[{"role":"user","content":"what is 2+3*8_3 ?"}]})



# for r in res["messages"]:
#     print(res)
#     print("\n")

print(res["messages"][-1].content)
```
