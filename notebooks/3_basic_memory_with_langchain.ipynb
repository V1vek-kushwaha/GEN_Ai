```python
from dotenv import load_dotenv
load_dotenv()
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")


promts =[
    {"role":"user","content":"hello my name is vivek kushwaha"},
    {"role":"user","content":"what is my name"}
]

res = llm.invoke(promts)
print(res.content)
```

## Manage History


```python
history = []

while True:
    quary = input("user:")
    
    if quary.lower() in ["exit","quit","bye"]:
        print("Good Bye ")

    history.append({"role":"user","content":quary})
    print("User " ,quary)

    res = llm.invoke(history)
    history.append({"role":"ai","content":res.content})

    print("AI ", res.content, "\n")

```
