## Connect with google Gen AI

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")


res = llm.invoke("who is the greatest king of india")
print(res.content)
```

## Connect with Open AI


```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-40")

res = llm.invoke("who is nareder modi")

print(res.content)
```

### Static Promts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

prompts = [
    ("system","you are a python developer"),  ## "system" ,"ai" , "user"
    ("user","how to sort array")
]
res = llm.invoke(prompts)
print(res.content)
```
